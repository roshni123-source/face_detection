```python
# import library
import cv2
# read image
img=cv2.imread("data\lena.png")
cv2.imshow("Lena",img)
cv2.waitKey(0)
```

```python
# convert into gray
g=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2.imshow('gray image',g)
cv2.waitKey(0)
```

```python
# haar cascade for reading dataset of faces
h=cv2.CascadeClassifier(r'C:\Users\HP\PycharmProjects\opencv\venv\haar_face.xml')
face=h.detectMultiScale(g,scaleFactor=1.1,minNeighbors=3)
print(f"Number of faces found = {len(face)}")
for (x,y,w,h) in face:
    cv2.rectangle(img, (x,y), (x+w,y+h), (0,255,0), thickness=2)
cv2.imshow('Detected Faces', img)
cv2.waitKey(0)
```
